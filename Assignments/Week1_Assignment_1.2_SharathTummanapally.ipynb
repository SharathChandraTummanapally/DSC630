```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Loading CSV data into data frame.
Covid_byState_df = pd.read_csv('Data/United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv')

# Checking the data.
Covid_byState_df.head()
```

This dataset contains daily numbers of confirmed and probable case and deaths reported to CDC by states over time.

### Create a histogram or bar graph from your data.

```python
# Plotting bar graph of the Top 10 states by COVID deaths.
Covid_byState_df.groupby('state')['conf_death'].sum().nlargest(10).plot(kind='bar')
plt.title('Top 10 states by COVID-19 deaths')           
plt.ylabel('')
plt.show()
```

### Create a boxplot from your data.

```python
# Create a boxplot of new cases.
plt.boxplot(Covid_byState_df['new_case'])
plt.title('New cases')
plt.show()
```

### Create a bivariate plot from your data.

```python
Covid_byState_df.groupby('state')[['tot_cases','tot_death']].sum()
```

```python
# Creating a scatter plot of Total cases vs Total deaths
Covid_byState_df.groupby('state')[['tot_cases','tot_death']].sum()
plt.scatter(Covid_byState_df['tot_cases'], Covid_byState_df['tot_death'])
plt.xlabel('Total Cases')
plt.ylabel('Total Deaths')
plt.title('Total cases vs Total Deaths')
plt.show()
```

### Create any additional visualizations that will help to answer the question(s) you want to answer.

```python
# Line plotting daily new cases.
Covid_byState_df['submission_date'] =  pd.to_datetime(Covid_byState_df['submission_date'])
Covid_byState_df.groupby('submission_date')['new_case'].sum().plot(kind='line',color = 'blue')
plt.title('Daily New Cases')          
plt.xlabel('Month')  
plt.ylabel('No. of Cases')
plt.show()
```

```python
# Line plotting daily new deaths.
Covid_byState_df['submission_date'] =  pd.to_datetime(Covid_byState_df['submission_date'])
Covid_byState_df.groupby('submission_date')['new_death'].sum().plot(kind='line',color = 'red')
plt.title('Daily New deaths')          
plt.xlabel('Month')  
plt.ylabel('No. of deaths')
plt.show()
```
